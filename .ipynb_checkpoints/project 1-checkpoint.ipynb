```python
corr_matrix = data.corr()
plt.figure(figsize=(15,6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='seismic', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
data .columns
```

```python
data.loc[:, 'target'] = data['target'].astype(str)
sns.countplot(x="sex", hue="target", data=data)
plt.xticks([0, 1], ['female', 'male'])
handles,labels = plt.gca().get_legend_handles_labels()
labels = ['negative' if label == '0' else 'positive' for label in labels]
plt.legend(handles, labels, title='target')
plt.show()
```

```python
age_cleaned = data['age'].replace([np.inf, -np.inf], np.nan)
sns.displot(data['age'])
plt.show
```

```python
sns.countplot(x='cp',data=data)
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-anginal pain","assymptomatic"])
plt.show()
```

```python
data.columns
```

```python
data.loc[:, 'target'] = data['target'].astype(str)
sns.countplot(x="cp", hue="target", data=data)
handles, labels = plt.gca().get_legend_handles_labels()
labels = ['negative' if label == '0' else 'positive' for label in labels]
plt.legend(handles, labels, title='target')
plt.show()
```

```python
data.replace([np.inf, -np.inf], np.nan, inplace=True)
g = sns.FacetGrid(data,hue="sex",aspect=4)
g.map(sns.kdeplot, 'trestbps', fill=True)
g.add_legend(title='Sex', labels=['male', 'female'])
plt.show()
```

```python
cate_val=[]
cont_val=[]
for column in data .columns:
 if data[column].nunique() <= 10:
    cate_val.append(column)
 else:
    cont_val.append(column)
print("Categorical variables:", cate_val)
print("Continuous variables:", cont_val)
```
